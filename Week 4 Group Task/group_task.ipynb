```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# logistic growth function: dA/dt = kA(1-A)
def dA_dt(t, A, k=1):
    return k * A * (1 - A)

# Initial condition
A0 = 0.5  # A(0)=0.5
t_span = (0, 10)   # time from t=0 to t=10
t_eval = np.linspace(0, 10, 200)  # for smooth plotting

# Solve ODE
sol = solve_ivp(dA_dt, t_span, [A0], t_eval=t_eval)

# Plot
plt.plot(sol.t, sol.y[0])
plt.xlabel("Years (t)")
plt.ylabel("A(t): Flower Coverage Fraction")
plt.title("Logistic Growth of Alien Flowers")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def dA_dt_season(t, A, k0=1, k1=2):
    k = k0 + k1 * np.sin(2 * np.pi * t)  # seasonal growth rate
    return k * A * (1 - A)

A0 = 0.5
t_span = (0, 10)
t_eval = np.linspace(0, 10, 500)

sol = solve_ivp(dA_dt_season, t_span, [A0], t_eval=t_eval)

plt.plot(sol.t, sol.y[0])
plt.xlabel("Years (t)")
plt.ylabel("A(t): Flower Coverage Fraction")
plt.title("Flower Growth with Seasonal Variation")
plt.grid(True)
plt.show()
```
